```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def reg1dim1(x, y):
    a = np.dot(x, y) / (x**2).sum()
    return a

x = np.array([1, 2, 4, 6, 7])
y = np.array([1, 3, 3, 5, 4])
a = reg1dim1(x, y)
```

```python
plt.scatter(x, y, color="k")
xmax = x.max()
plt.plot([0, xmax], [0, a*xmax], color='k')
plt.show()
```

```python
def reg1dim2(x, y):
    n = len(x)
    a = ((np.dot(x, y) - y.sum() * x.sum() / n)) / \
        ((x**2).sum() - x.sum()**2 / n )
    b = (y - a * x).sum() / n
    return a, b
```

```python
x = np.array([1, 2, 4, 6, 7])
y = np.array([1, 3, 3, 5, 4])
a, b = reg1dim2(x, y)
```

```python
plt.scatter(x, y, color='k')
xmax = x.max()
plt.plot([0, xmax], [0, xmax*a+b], color='k')
plt.show()
```
